```python
import tensorflow as tf
import numpy as np
```

```python
a = tf.constant(3.0)
b = tf.constant(4.0)
c = tf.sqrt(a**2 + b**2)

print('c:', c)
print(f'c: {c}')
print('c in numpy:', c.numpy())
```

```python
# vector multiplication
a = tf.constant([1,2,3])
b = tf.constant([4,5,6])
c = tf.tensordot(a, b, axes=[0,0])

print('c:', c)
print(f'c: {c}')
print('c in numpy:', a.numpy().dot(b.numpy()))
```

```python
# matrix multiplication
A0 = np.random.randn(3,3)
b0 = np.random.randn(3,1)
A = tf.constant(A0)
b = tf.constant(b0)
c = tf.matmul(A, b)

print('c:', c)
print(f'c: {c}')
print('c:', c.numpy())
print('c in numpy:', A0.dot(b0))
```

```python
# broadcasting
a = tf.constant([[1,2],[3,4]])
b = tf.constant(1)
c = a + b

print('c:', c)
print(f'c: {c}')
print('c in numpy:', a.numpy() + b.numpy())
```

```python
# elementwise multiplication
a = tf.constant([[1,2],[3,4]])
b = tf.constant([[2,3],[4,5]])
c = a * b

print('c:', c)
print(f'c: {c}')
print('c in numpy:', a.numpy() * b.numpy())
```
