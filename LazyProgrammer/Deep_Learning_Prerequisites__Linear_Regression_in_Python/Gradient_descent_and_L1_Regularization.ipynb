```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent_step(df, x, step_size):
  return x - step_size * df(x)

def gradient_descent(f, df, x0, step_size, numer_of_steps, treshold=None, should_trace_steps=None, should_plot=False):
  x = x0
  
  if (should_trace_steps is not None):
    print(f'Step 0. Starting point is {x0} and initial function value is {f(x0)}')
  
  if (should_plot):
    steps_hist = [0]
    f_hist = [f(x0)]

  for step in range(numer_of_steps):
    x_new = gradient_descent_step(df, x, step_size)
    f_new = f(x_new)
    if (should_trace_steps is not None):
      if ((step+1) % should_trace_steps == 0):
        print(f'Step {step+1}. At {x_new} function value is {f_new}')
    
    if (should_plot):
      steps_hist.append(step+1)
      f_hist.append(f_new)
    
    if (treshold is not None):
      change = np.abs(f_new - f(x))
      if (change < treshold):
        print(f'Early stop after {step+1} steps because change in step {change} is smaller than treshold {treshold}')
        x = x_new 
        break
   
    x = x_new 


  if (should_trace_steps is not None):
    print(f'Loop finished with x at {x} and function value is {f(x)}')
  
  if (should_plot):
    plt.figure(figsize=(20,10))
    plt.plot(steps_hist, f_hist);
    plt.title('Cost fuction history');

  return x
```

```python
fun = lambda x: (x ** 4) - x ** 2
dfun = lambda x: 4 * (x ** 3) - 2 * x

_ = gradient_descent(fun, dfun, x0=5, step_size=0.001, numer_of_steps=100, treshold=1, should_trace_steps=1, should_plot=True)
```

# Gradient descent for Linear Regression

```python
## [1, X, X**2] features example
```

```python
N = 100
D = 3

t = np.linspace(0, 10, N)
t2 = t ** 2

X = np.vstack((np.ones(N), t, t2)).T
print(f'X shape: {X.shape}')
Y = 0.5 * X[:,1] + np.random.randn(N)
Y[-1] +=30
Y[-2] +=30

w0 = np.random.randn(D) / np.sqrt(D)
cost = lambda  w: (X.dot(w) - Y).dot(X.dot(w) - Y)
dcost = lambda  w: X.T.dot(X.dot(w) - Y)

_ = gradient_descent(cost, dcost, x0=w0, step_size=0.000001, numer_of_steps=1000, treshold=0.001, should_trace_steps=20, should_plot=True)
```

## L1 regularization

```python
N = 50
D = 50

X = (np.random.random((N,D)) - 0.5) * 10
true_w = np.array([1, 0.5, -0.5] + [0]*(D-3))
Y = X.dot(true_w) + np.random.randn(N) * 0.5

w0 = np.random.randn(D) / np.sqrt(D)
p_lambda = 10.0

l1_cost = lambda  w: (Y - X.dot(w)).dot(Y - X.dot(w)) + p_lambda * np.sqrt(w.dot(w))
dl1_cost = lambda  w: X.T.dot(X.dot(w) - Y) + p_lambda * np.sign(w)

w = gradient_descent(l1_cost, dl1_cost, x0=w0, step_size=0.001, numer_of_steps=500, treshold=None, should_trace_steps=100, should_plot=True)

plt.figure(figsize=(20,10))
plt.plot(X.dot(w), label='prediction');
plt.plot(Y, label='target');
plt.legend()
plt.title('Predictions vs targets')

plt.figure(figsize=(20,10))
plt.plot(true_w, label='true_w');
plt.plot(w, label='MAP estimated w');
plt.legend()
plt.title('Estimated weights')
plt.show()

print('Estimated weights:', w)
```
